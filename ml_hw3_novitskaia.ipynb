```python
alphabet_dic = {}
for i in range(len(alphabet)):
    alphabet_dic[alphabet[i]] = i
```

```python
M = np.zeros((len(alphabet), len(alphabet)))
```

```python
for key in bigrams.keys():
    M[alphabet_dic[key[0]]][alphabet_dic[key[1]]] = bigrams[key]
```

```python
def score(text, decoding, M):
    score = 0.
    for i in range(len(text) - 2):
        l_1 = text[i:i+2][0]
        l_2 = text[i:i+2][1]
        score_cur = M[alphabet_dic[decoding[l_1]]][alphabet_dic[decoding[l_2]]]
        score += np.log(score_cur)
    return score
```

```python
def change_decoder(alphabet, decoding_alphabet):
    idx_1 = np.random.randint(0, len(alphabet))
    idx_2 = np.random.randint(0, len(alphabet))
    while idx_1 == idx_2:
        idx_2 = np.random.randint(0, len(alphabet))
    decoding_alphabet[idx_1], decoding_alphabet[idx_2] = decoding_alphabet[idx_2], decoding_alphabet[idx_1]
    decoding = {}
    for i in range(len(alphabet)):
        decoding[decoding_alphabet[i]] = alphabet[i]
    return decoding, decoding_alphabet
```

```python
decoding_alphabet
```

```python
for epoch in range(20000):
    score_1 = score(test_coding, decoding_3, M)
    decoding_3_next = {}
    decoding_alphabet_next = []
    decoding_3_next, decoding_alphabet_next = change_decoder(alphabet.copy(), decoding_alphabet.copy())
    score_2 = score(test_coding, decoding_3_next, M)
    if score_1 < score_2:
        decoding_3 = decoding_3_next.copy()
        decoding_alphabet = decoding_alphabet_next.copy()
    else:
        p = np.exp(score_2 - score_1)
        x = np.random.choice(2, 1, p=[1 - p, p])[0]
        if x == 1:
            decoding_3 = decoding_3_next.copy()
            decoding_alphabet = decoding_alphabet_next.copy()
    if epoch % 1000 == 0:
        score_1 = score(test_coding, decoding_3, M)
        test_decoding = ''
        for s in test_coding:
            test_decoding += decoding_3[s]
        print('Current score = ' + str(score_1))
        print(test_decoding)
```
